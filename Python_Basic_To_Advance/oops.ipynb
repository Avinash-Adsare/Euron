```python
## Class
class person:
    # constructor

    def __init__(self,name,age):
        self.name = name
        self.age = age
        print(f"to check values are initialised or not {self.name} and {self.age}")

        
    def get_name(self):
        return self.name
    
    def get_age(self):
        return self.age
    
    
```

## Class method

- handle class attribute
- @classmethod
- cls(used)
- access by clssname
- modify

```python
class netflix:

    new_show = "part4"

    def __init__(self,mail,password):
        self.mail = mail
        self.password = password
        self.service = "free"
        self.amount = 0

        # print(f"this is my mail {self.mail} and this is {self.password} and service {self.service} and amount {self.amount}")

    def get_password(self):
        return self.password


    @classmethod    ## class method
    def get_new_show(cls):
        return cls.new_show
    
    #to change/ modify
    @classmethod
    def set_new_show(cls,show_name):
        cls.new_show = show_name
```

```python
netflix.get_new_show()
```

```python
myuser= netflix("prince",23)
myuser.get_new_show()
```

```python
netflix.set_new_show("part6")
print(netflix.get_new_show())
```

# static method

```python
class mymath:
    @staticmethod
    def add(x,y):  ## utility function
        return x+y
    
mymath.add(3,4)
```

```python
class netflix:
    new_show = "part5"

    def __init__(self,mail,password):
        self.mail = mail
        self.password = password
        self.service = "free"
        self.amount = 0

        print(f"this is my mail {self.mail} and this is {self.password} and service {self.service} and amount {self.amount}")

    def get_password(self):
        return self.password
    
    @staticmethod
    def get_new_show():
        return netflix.new_show
```
