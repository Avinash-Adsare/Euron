## Set Compreshension

- set comprehension is a consise way to create sets in python. It follows a similar structure to list compreshension,but it generates a set instead of a list

```python
# value square

for i in range(10):
    print(i**2)
```

```python
square = {i**2 for i in range(10) if i%2==0 }
square
```

```python
data = [1,2,3,4,5,6,7,8,9]
even = {i **2 for i in data if i%2 == 0}
even
```

```python
words = ["apple","banana","cherry","date"]

{len(word) for word in words}
```

```python
# Q.Remove duplicates from an array

l = [1,2,3,6,4,5,7,9,4,554,758,46,9]

set(l)
```

```python
 # Q.Remove duplicates from an array by fun

def remove_dup(l):
    return list(set(l))
l = [1,2,3,6,4,5,7,9,4,554,758,46,9]

remove_dup(l)
```
