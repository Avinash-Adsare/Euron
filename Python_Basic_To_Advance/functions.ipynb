# Nested functions

```python
def outer_fun(outer):

    def inner_fun(inner):
        result = outer+inner
        return result
    result_inner= inner_fun(5)

    return result_inner

result_outer = outer_fun(10)
print(result_outer)
```

```python
def sum_of_cubes(a,b):

    def cube(x):
        return x*x*x
        

    def add(x,y):
        return x+y
    
    return add(cube(a),cube(b))



result = sum_of_cubes(2,3)
print(result)
```

```python
num = [1,6,2,155,456,524] ## custom loop
def my_loop(num):
    iterator =  iter(num)
    print(next(iterator))
    while True:
        try:
            print(next(iterator))
        except StopIteration:
            break

my_loop(num)
```

## Custom Range Function

```python
## khud ka range fun

class myrange:

    def __init__(self,start,stop,step):
        self.start = start
        self.stop = stop
        self.step = step
    def __iter__(self):
        return self
    def __next__(self):
        if self.start < self.stop:
            value =self.start
            self.start = self.start + self.step
            return value
        else:
            raise StopIteration
```

```python
r = myrange(1,10,2)
for i in r:
    print(i)
```

```python
# generator fun for fibo
# yeild keyword make generator fun
def fibo(n):
    a,b = 0,1
    while a < n :
        yield a
        a,b = b,a+b

for i in fibo(1000):
    print(i)
```

# Decorators :
- dusre fun  ek fun ke andar

decoration
function

```python
def deco(func):
    def fun():
        print("#######")
        func()
        print("#######")

    return fun
```

```python
@deco
def hello():
    print("hello, People")

hello()
```

```python
import time

def time_deco(func):
    def kaam():
        start_time = time.time()
        result = func()
        end_time = time.time()
        print(f"execution time {end_time-start_time:5f}")

        return result
    return kaam
```

```python
@time_deco
def base():
    print("Start execution")
    time.sleep(2)
base()
```
