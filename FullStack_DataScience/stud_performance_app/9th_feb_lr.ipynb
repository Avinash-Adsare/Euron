```python
from sklearn.preprocessing import LabelEncoder,StandardScaler  #labelencoder means change the labels
```

```python
df["Extracurricular Activities"] # select because it is object type and convert it 1/0 format
```

```python
le = LabelEncoder()
df["Extracurricular Activities"] = le.fit_transform(df["Extracurricular Activities"]) # encoding
```

```python
# update the entire df
df
```

# once df is numerical data then create X and Y.

```python
# y = performace index

# x creation
x = df[["Hours Studied","Previous Scores","Extracurricular Activities","Sleep Hours","Sample Question Papers Practiced"]]
```

```python
x
```

```python
y = df["Performance Index"]
```

```python
y
```

```python
# split the data in test and train

from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=.2) #test data is 20 % , 80 % to train
```

```python
x_train # 80 % data included to train
```

```python
x_test #20 % data is used for testing
```

```python
# z-score means - Avg / std deviation
```

```python
scaler = StandardScaler()   # standardize the data
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)
```

```python
# linear regression training

from sklearn.linear_model import LinearRegression
import pickle #pickel is physical file we created
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
# train the model
model = LinearRegression()
model.fit(x_train_scaled,y_train)
```

```python
y_pred = model.predict(x_test_scaled)
```

```python
x_test_scaled
```

```python
# we can pass one record also and test it

model.predict([[ 0.80712124, -0.53764053,  1.03149575, -0.88839972,  0.1345192 ]])
```

```python
mean_squared_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred) # r2 checks the model accuracy
```

```python
# create a physical file
import pickle
with open("Student_lr_final_model.pkl",'wb') as file:
    pickle.dump((model,scaler,le),file)
```
