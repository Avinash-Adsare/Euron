Beautiful Soup=> Gives control over to html.

Beautiful Soup is a Python package for parsing HTML and XML documents, including those with malformed markup. It creates a parse tree for documents that can be used to extract data from HTML, which is useful for web scraping.

Web extracting . Extract any data from web 

```python
from bs4 import BeautifulSoup
```

```python
html_data = """
<html> <head><title> The Title </title></head>
<body>
<h2>Hello,World!</h2>
<h1> This is my first html tag my name is Avinsh </h1>
<h1> This is my first html tag my name is Avinsh </h1>
<h1> This is my first html tag my name is Avinsh </h1>
<h1> This is my first html tag my name is Avinsh </h1>
<h1> This is my first html tag my name is Avinsh </h1>
<h1> This is my first html tag my name is Avinsh </h1>
<p>This is a paragraph. I am trying new things</p></body></html>
"""
```

```python
data = BeautifulSoup(html_data,'html.parser') ## beautifulsoup give control on html ,xml data
```

```python
for item in data.find_all('h1'):  ## sare h1 calling with the help of loop
    print(item)
```
